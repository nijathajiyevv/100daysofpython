```python
import random

# Initialize the deck with card values
cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]

# betting
starting_amount = 1000
print("You have $" + str(starting_amount)+" starting amount")
bet = int(input("How much do you want to bet? "))
if bet > starting_amount:
    print("You don't have enough money")
    exit()



# Function to calculate the value of a hand
def calculate_hand_value(hand):
    value = sum(hand)
    # Check for aces and adjust the value if necessary
    if value > 21 and 11 in hand:
        hand.remove(11)
        hand.append(1)
        value = sum(hand)
    return value

# Function to display a hand
def display_hand(hand):
    for card in hand:
        print(card, end=' ')
    print()

# Shuffle the deck
random.shuffle(cards)

# Deal initial cards
player_hand = [cards.pop(), cards.pop()]
dealer_hand = [cards.pop(), cards.pop()]

# Main game loop
while True:
    player_score = calculate_hand_value(player_hand)
    dealer_score = calculate_hand_value(dealer_hand)

    print("\nPlayer's hand:")
    display_hand(player_hand)
    print(f"Player's score: {player_score}")

    print("\nDealer's hand:")
    print(dealer_hand[0])

    # Check for Blackjack
    if player_score == 21 and len(player_hand) == 2:
        print("Blackjack! Player wins!")
        # delete bet money from starting amount
        starting_amount = starting_amount - bet
        break
    elif dealer_score == 21 and len(dealer_hand) == 2:
        print("Blackjack! Dealer wins!")
        # add bet money to starting amount
        starting_amount = starting_amount + bet
        break

    # Player's turn
    while True:
        choice = input("\nDo you want to 'hit' or 'stand'? ").strip().lower()
        if choice == 'hit':
            player_hand.append(cards.pop())
            player_score = calculate_hand_value(player_hand)
            print("\nPlayer's hand:")
            display_hand(player_hand)
            print(f"Player's score: {player_score}")
            if player_score > 21:
                print("Player busts! Dealer wins!")
                # add bet money to starting amount
                starting_amount = starting_amount + bet
                break
        elif choice == 'stand':
            break
        else:
            print("Invalid choice. Please enter 'hit' or 'stand'.")

    if player_score > 21:
        break

    # Dealer's turn
    while dealer_score < 17:
        dealer_hand.append(cards.pop())
        dealer_score = calculate_hand_value(dealer_hand)

    print("\nDealer's hand:")
    display_hand(dealer_hand)
    print(f"Dealer's score: {dealer_score}")

    if dealer_score > 21:
        print("Dealer busts! Player wins!")
        starting_amount = starting_amount - bet
    elif dealer_score > player_score:
        print("Dealer wins!")
        starting_amount = starting_amount + bet
    elif player_score > dealer_score:
        print("Player wins!")
        starting_amount = starting_amount - bet
    else:
        print("It's a tie!")
    print("You have $" + str(starting_amount)+" remaining")
    break

# End of the game
```
