```python
# import packages
import random
import timer

# Welcome user

print("Welcome to the word guessing game!")

# word list

capitals = ["baku", "tbilisi", "sofia", "warsaw", "berlin", "paris", "london", 
             "madrid", "rome", "vienna", "milan", "budapest", "prague", "brussels", "amsterdam", 
             "stockholm", "athens", "dublin", "oslo", "helsinki"]

vegetables = ["carrot", "broccoli", "spinach", "tomato", 
              "cucumber", "pepper", "onion", "zucchini", 
              "eggplant", "celery", "potato", "lettuce"]

fruits = ["apple", "banana", "orange", "grape", "strawberry", 
          "pineapple", "watermelon", "kiwi", "mango", 
          "blueberry", "pear", "cherry"]

home_stuff = ["couch", "table", "chair", "lamp", "bed", 
              "mirror", "clock", "sofa", "bookshelf", 
              "vase", "curtains", "rug"]

brands = ["apple", "nike", "coca-cola", "amazon", "adidas", 
          "microsoft", "google", "tesla", "pepsi", "samsung", 
          "bmw", "sony"]

# ask for list selection

listselection = input("Which list would you like to play? (capitals, vegetables, fruits, homestuff, brands) ")


if listselection == "capitals":
    word_list = capitals
elif listselection == "vegetables":
    word_list = vegetables
elif listselection == "fruits":
    word_list = fruits
elif listselection == "homestuff":
    word_list = home_stuff
elif listselection == "brands":
    word_list = brands

# random word selection
listselection = word_list

chosen_word = random.choice(word_list)
word_length = len(chosen_word)
print(f"You have {word_length} lives to guess the word.")


end_of_game = False
lives = 6


#Create blanks
display = []
for _ in range(word_length):
    display += "_"

while not end_of_game:
    guess = input("Guess a letter: ").lower()
    if guess in display:
        print(f"You've already guessed {guess}")

    #Check guessed letter
    for position in range(word_length):
        letter = chosen_word[position]
        if letter == guess:
            display[position] = letter

    # display guessed history
    guessed_letters = []
    for letter in display:
        if letter != "_":
            guessed_letters.append(letter)
            
    print(f"You've guessed these letters: {', '.join(guessed_letters)}")

    #Check if user is wrong.
    if guess not in chosen_word:
        #TODO-5: - If the letter is not in the chosen_word, print out the letter and let them know it's not in the word.
        print(f"You guessed {guess}, that's not in the word. You lose a life.")
        
        lives -= 1
        print(f"You have {lives} lives left.")
        if lives == 0:
            end_of_game = True
            print("You lose.")
            print(f"The correct word is *{chosen_word}*")
        

    #Join all the elements in the list and turn it into a String.
    print(f"{' '.join(display)}")

    #Check if user has got all letters.
    if "_" not in display:
        end_of_game = True
        print("You win.")
```
